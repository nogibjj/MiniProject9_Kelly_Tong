```python
# NBVAL_IGNORE_OUTPUT
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import numpy as np
```

Loading Package

```python
# NBVAL_IGNORE_OUTPUT
# Load the 'Auto' dataset
path = "https://raw.githubusercontent.com/nogibjj/MiniProject9_Kelly_Tong/main/test_Auto.csv"
df = pd.read_csv(path)
df.head()
```

### Functions for Data Manipulations

```python
# NBVAL_IGNORE_OUTPUT
def desc_stats(data):
    # Convert the dataset to a Pandas DataFrame
    Auto = pd.DataFrame(data)

    # Display a brief summary of the dataset
    print(Auto.info())

    # Glimpse the first few rows of the dataset
    print(Auto.head())

    # Summary of the numeric variables
    print(Auto.describe())

    return None
```

```python
# NBVAL_IGNORE_OUTPUT
def box_visual(data):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        sns.set_theme(style="ticks", palette="pastel")
        sns.boxplot(x="mpg", y="weight", palette="Blues", data=data)
        plt.show()
```

### Describe Data

```python
# NBVAL_IGNORE_OUTPUT
desc_stats(df)
```

### Data Visualization

```python
# NBVAL_IGNORE_OUTPUT
box_visual(df)
```

### More Data Manipulation

```python
# NBVAL_IGNORE_OUTPUT
# Generating random dataset with same number of rows as the cars dataset
num_rows = len(df)

# Random data generation
np.random.seed(42)  # for reproducibility
random_data = {
    "Random_Value1": np.random.randint(
        1, 101, num_rows
    ),  # Random integers between 1 and 100
    "Random_Value2": np.random.rand(num_rows) * 100,  # Random floats between 0 and 100
    "Random_Category": np.random.choice(
        ["A", "B", "C", "D"], num_rows
    ),  # Random selection from categories A, B, C, D
}

random_df = pd.DataFrame(random_data)

random_df.head()
```

Merging Two Datasets

```python
# NBVAL_IGNORE_OUTPUT
merged_df = pd.concat([df, random_df], axis=1)

merged_df.head()
```
